# Comparing different spots on the surface of beta Pic b using the starry library

This notebook is used as a tutorial and a playground to test the capabilities and the results of the `starry` library. The goal is to test the effect of the existence of spots on the surface of a planet on the light curve of the planet. The planet we are going to use is beta Pic b, and we are going to compare the light curves of different spots on the surface of the planet.

```python
import numpy as np
import matplotlib.pyplot as plt
import starry
import astropy.units as u
from constants import *

starry.config.lazy = False
starry.config.quiet = True
```

Create arrays of angle and time for use throughout the notebook.

```python
beta_mass=11.9 #in u.M_jup
beta_rad=1.65  #in u.R_jup
theta = np.linspace(-180, 180, 1000)
time_ros = np.linspace(0,BETA_DAY,len(theta))/(60*60)
```

## Testing different functionalities of the starry library

### Testing the effects of limb-darkening

```python
A = starry.Primary(
    starry.Map(ydeg=10, udeg=2, rv=True, amp=1, veq=BETA_VEQ, alpha=0),
    r=beta_rad,
    m=beta_mass,
    length_unit=u.Rjup,
    mass_unit=u.Mjup,
)

map_beta=A.map
map_beta[1] = 0.5
map_beta[2] = 0.25

map_beta.add_spot(-0.015, sigma=0.01, lat=30, lon=0)

flux_1_ros = map_beta.flux(theta=theta)
rv_1_ros = map_beta.rv(theta=theta)
map_beta.show(rv=False,theta=np.linspace(0,360,50))
```

```python
A = starry.Primary(
    starry.Map(ydeg=10, udeg=2, rv=True, amp=1, veq=BETA_VEQ, alpha=0),
    r=beta_rad,
    m=beta_mass,
    length_unit=u.Rjup,
    mass_unit=u.Mjup,
)

map_beta=A.map
#map_beta[1] = 0.5
#map_beta[2] = 0.25

map_beta.add_spot(-0.015, sigma=0.01, lat=30, lon=0)

flux_2_ros = map_beta.flux(theta=theta)
rv_2_ros = map_beta.rv(theta=theta)
map_beta.show(rv=False,theta=np.linspace(0,360,50))
```

```python
# Plot the radial velocity
plt.figure(figsize=(12, 5))
plt.title('Comparing with or without ydeg')
plt.plot(time_ros, rv_1_ros,label='with ydeg')
plt.plot(time_ros, rv_2_ros,label='without ydeg')
plt.legend()
plt.xlabel("Time [hours]", fontsize=24)
plt.ylabel("Radial velocity [m/s]", fontsize=24)
plt.grid()
plt.show()
```
